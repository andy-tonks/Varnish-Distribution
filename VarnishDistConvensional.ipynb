```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('6Months.csv',encoding='latin1')
df.head()
```

```python
df.drop(df[df['Esp. de Topo OV'] < 0].index,inplace=True)
df.drop(df[df['Esp. de Fundo OV'] < 0].index,inplace=True)
df.drop(df[df['Esp. de Topo OV'] > 7].index,inplace=True)
df.drop(df[df['Esp. de Meio OV'] < 0].index,inplace=True)
df.describe()
```

```python
df['Esp. de Fundo OV'].plot()
```

```python
df['Produto'].value_counts()
```

```python
grps = df.groupby('Produto')
Con_473= grps.get_group('OV - 473ml - Convencional')

print(Con_473.describe())
```

```python
fig,axes = plt.subplots(1,3,figsize=(15,4))
plt.title('473ml - Convensional')
sns.histplot(Con_473['Esp. de Topo OV'],bins=20,ax=axes[0])
sns.histplot(Con_473['Esp. de Meio OV'],bins=20,ax=axes[1])
sns.histplot(Con_473['Esp. de Fundo OV'],bins=20,ax=axes[2])
```

```python
fig,axes = plt.subplots(3,1,figsize=(20,15))
plt.title('473ml -Convensional')

for ax in axes.flat:
    ax.axline((0,1.2), slope=0, color='r', linewidth=2, linestyle='-')
    ax.axline((0,3.6), slope=0, color='r', linewidth=2, linestyle='-')
    ax.axline((0,2.4), slope=0, color='g', linewidth=2, linestyle='--')
    ax.text(s='Min - 1.2',y=1.3,x=25000)
    ax.text(s='Max - 3.6',y=3.65,x=25000)
    ax.text(s='Target - 2.4',y=2.5,x=25000)

    

sns.scatterplot(y=Con_473['Esp. de Topo OV'],x= Con_473.index,hue=Con_473['Processo'],ax=axes[0])
sns.scatterplot(y=Con_473['Esp. de Meio OV'],x= Con_473.index,hue=Con_473['Processo'],ax=axes[1])
sns.scatterplot(y=Con_473['Esp. de Fundo OV'],x= Con_473.index,hue=Con_473['Processo'],ax=axes[2])

axes[0].legend(loc=(1.01,0.5)) 
axes[1].legend(loc=(1.01,0.5)) 
axes[2].legend(loc=(1.01,0.5)) 


# plt.figure(figsize=(20,4))
# plt.title('473ml -Convensional')
# sns.scatterplot(y=Con_473['Esp. de Meio OV'],x= Con_473.index,hue=Con_473['Processo'])
# plt.axhline(1.2,color='r')
# plt.axhline(3.92,color='r')
# plt.axhline(2.56,color='g')
# plt.text(s='Min - 1.2',y=1.21,x=5600)
# plt.text(s='Max - 3.92',y=3.93,x=5600)
# plt.text(s='Target - 2.56',y=2.57,x=5600)


#limit total devPad 4.
```
